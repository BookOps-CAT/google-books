```python
gdf = pd.read_csv(gfh, delimiter="\t", low_memory=False, header=0, dtype={"Barcode": str}, names=[
    "Barcode", "Checkin_date", "State", "Viewability", "Condtions", "Scannable", "Optout", "Tagging", "Audit", "Material_err",
    "Overall_err", "Scan_date", "Proc_date_google", "Anlyzed_date", "Converted_date", "Allow_download_date", "Viewability_date", "Src_library_bibkey", "Rubbish", "Downloaded_date",
    "Claimed", "Ocr_gtd_score", "OCR_analysis_score", "Digitization_method", "Ocred_date"
])
```
